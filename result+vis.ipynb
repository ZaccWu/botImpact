```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# t1 pos
plt.figure(figsize=(2,10))

treat_naive = 0.2052
control_naive = 0.1487
treat_AdvG = 0.2326
control_AdvG = 0.1475

plt.plot([1, 1], [treat_naive, control_naive], color='b', label='Naive', linewidth=4, zorder=1)
plt.plot([1.2, 1.2], [treat_AdvG, control_AdvG], color='r', label='AdvG', linewidth=4,  zorder=1)

plt.scatter(1, treat_naive, color='b', marker = 's', s=300, linewidths=1.5, edgecolors='black', zorder=2)
plt.scatter(1, control_naive, color='b', marker = 'o', s=300, linewidths=1.5, edgecolors='black', zorder=2)

plt.text(1, treat_naive+0.006, treat_naive, ha='center', va='bottom',fontsize=16)
plt.text(1, control_naive-0.008, control_naive, ha='center', va='top',fontsize=16)

plt.scatter(1.2, treat_AdvG, color='r', marker = 's', s=300, linewidths=1.5, edgecolors='black', zorder=2)
plt.scatter(1.2, control_AdvG, color='r', marker = 'o', s=300, linewidths=1.5, edgecolors='black', zorder=2)

plt.text(1.2, treat_AdvG+0.006, treat_AdvG, ha='center', va='bottom',fontsize=16)
plt.text(1.2, control_AdvG-0.015, control_AdvG, ha='center', va='top',fontsize=16)

plt.tick_params(labelsize=16)
plt.ylim(0.05, 0.35)
plt.xlim(0.8, 1.4)
plt.xticks([1.1],[''])
plt.savefig('estimate_t1_pos.jpg', bbox_inches = 'tight', dpi=500)
plt.show()
```
