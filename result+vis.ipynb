```python
# t1 pos
fig = plt.figure(figsize=(8,5))

label = ['t1_pos', 't2_pos', 't3_pos', 't1_neg', 't2_neg', 't3_neg']
aucH = [0.8289, 0.7842, 0.8805, 0.8648, 0.8973, 0.9231]
aucB = [0.8105, 0.7094, 0.7924, 0.8096, 0.7991, 0.7791]

apH = [0.8325, 0.8225, 0.9003, 0.8564, 0.9135, 0.9346]
apB = [0.8110, 0.7916, 0.8425, 0.8049, 0.8160, 0.8277]

change = [0.0286, 0.0386, 0.0847, 0.0678, 0.1589, 0.1684]

axis1 = fig.add_axes((0.1, 0.1, 0.8, 0.8))
axis2 = axis1.twinx()

def plot_res(pos, id):
    axis2.scatter(pos-0.15, aucH[id], color='r', marker = 'o', s=200, linewidths=1.5, edgecolors='black', zorder=3)
    axis2.scatter(pos-0.15, aucB[id], color='r', marker = 's', s=200, linewidths=1.5, edgecolors='black', zorder=3)
    axis2.plot([pos-0.15, pos-0.15], [aucH[id], aucB[id]], color='r', label='Naive', linewidth=4, zorder=2)

    axis2.scatter(pos+0.15, apH[id], color='g', marker = 'o', s=200, linewidths=1.5, edgecolors='black', zorder=3)
    axis2.scatter(pos+0.15, apB[id], color='g', marker = 's', s=200, linewidths=1.5, edgecolors='black', zorder=3)
    axis2.plot([pos+0.15, pos+0.15], [apH[id], apB[id]], color='g', label='Naive', linewidth=4, zorder=2)

    axis1.bar(pos, height=change[id], width=0.3, color='gold', edgecolor='black', linewidth= 0.5)

for i in range(len(aucH)):
    plot_res(i+1, i)

axis1.set_ylim(0., 0.2)
axis2.set_ylim(0.6, 1)

axis1.tick_params(labelsize=16)
axis2.tick_params(labelsize=16)
plt.xlim(0, 7)
plt.xticks([1, 2, 3, 4, 5, 6],label,size=16)
plt.savefig('gcn_pred.jpg', bbox_inches = 'tight', dpi=500)
plt.show()
```

```python
# t1 pos
fig = plt.figure(figsize=(5,5))

label = ['t1_pos', 't2_pos', 't3_pos', 't1_neg', 't2_neg', 't3_neg']
aucH = [0.8289, 0.7842, 0.8805, 0.8648, 0.8973, 0.9231]
aucB = [0.8105, 0.7094, 0.7924, 0.8096, 0.7991, 0.7791]
change = [0.0286, 0.0386, 0.0847, 0.0678, 0.1589, 0.1684]

axis1 = fig.add_axes((0.1, 0.1, 0.8, 0.8))
axis2 = axis1.twinx()
axis2.errorbar(label, aucH, label='Human AUC', alpha=0.7, linewidth = 2, marker = 'o', elinewidth=0.5, capsize=1, capthick=2, ms=11, mec='black',
             color='r', zorder=2)
axis2.errorbar(label, aucB, label='Bot AUC', alpha=0.7, linewidth = 2, marker = 's', elinewidth=0.5, capsize=1, capthick=2, ms=11, mec='black',
             color='g', zorder=2)
axis1.bar(label, height=change, width=0.3, color='snow', hatch='//', edgecolor='black', linewidth= 0.3)
axis1.set_ylim(0., 0.2)
axis2.set_ylim(0.6, 1)
axis1.tick_params(labelsize=16)
axis2.tick_params(labelsize=16)

plt.legend(fontsize=18)
plt.savefig('gcn_auc.jpg', bbox_inches = 'tight', dpi=500)
plt.show()
```
