```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# t1 pos
plt.figure(figsize=(2,10))

treat_naive = 0.2052
control_naive = 0.1487
treat_AdvG = 0.2123
control_AdvG = 0.1106

plt.plot([1, 1], [treat_naive, control_naive], color='b', label='Naive', linewidth=4, zorder=1)
plt.plot([1.2, 1.2], [treat_AdvG, control_AdvG], color='r', label='AdvG', linewidth=4,  zorder=1)

plt.scatter(1, treat_naive, color='b', marker = 's', s=300, linewidths=1.5, edgecolors='black', zorder=2)
plt.scatter(1, control_naive, color='b', marker = 'o', s=300, linewidths=1.5, edgecolors='black', zorder=2)

plt.text(1, treat_naive+0.006, treat_naive, ha='center', va='bottom',fontsize=16)
plt.text(1, control_naive-0.008, control_naive, ha='center', va='top',fontsize=16)

plt.scatter(1.2, treat_AdvG, color='r', marker = 's', s=300, linewidths=1.5, edgecolors='black', zorder=2)
plt.scatter(1.2, control_AdvG, color='r', marker = 'o', s=300, linewidths=1.5, edgecolors='black', zorder=2)

plt.text(1.2, treat_AdvG+0.006, treat_AdvG, ha='center', va='bottom',fontsize=16)
plt.text(1.2, control_AdvG-0.006, control_AdvG, ha='center', va='top',fontsize=16)

plt.tick_params(labelsize=16)
plt.ylim(0.05, 0.25)
plt.xlim(0.8, 1.4)
plt.xticks([1.1],[''])
plt.savefig('estimate_t1_pos.jpg', bbox_inches = 'tight', dpi=500)
plt.show()
```

```python
# t1 pos
fig = plt.figure(figsize=(5,5))

label = ['t1_pos', 't1_neg', 't2_pos', 't2_neg', 't3_pos', 't3_neg']
aucH = [0.9261, 0.8570, 0.8318, 0.9225, 0.9023, 0.8709]
aucB = [0.8777, 0.8400, 0.7875, 0.8625, 0.8763, 0.8361]
change = [0.0452, 0.0365, 0.0426, 0.0519, 0.0270, 0.0458]

axis1 = fig.add_axes((0.1, 0.1, 0.8, 0.8))
axis2 = axis1.twinx()
axis2.errorbar(label, aucH, label='Human AUC', alpha=0.7, linewidth = 2, marker = 'o', elinewidth=0.5, capsize=1, capthick=2, ms=11, mec='black',
             color='r', zorder=2)
axis2.errorbar(label, aucB, label='Bot AUC', alpha=0.7, linewidth = 2, marker = 's', elinewidth=0.5, capsize=1, capthick=2, ms=11, mec='black',
             color='g', zorder=2)
axis1.bar(label, height=change, width=0.3, color='snow', hatch='//', edgecolor='black', linewidth= 0.3)
axis1.set_ylim(0., 0.1)
axis2.set_ylim(0.75, 0.95)
axis1.tick_params(labelsize=16)
axis2.tick_params(labelsize=16)

plt.legend(fontsize=18)
plt.savefig('gcn_auc.jpg', bbox_inches = 'tight', dpi=500)
plt.show()
```
