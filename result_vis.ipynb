```python
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
```

```python
# t1 pos
plt.figure(figsize=(5,5))

label = ['t1_pos', 't1_neg', 't2_pos', 't2_neg', 't3_pos', 't3_neg']
aucH = [0.9261, 0.8570, 0.8318, 0.9225, 0.9023, 0.8709]
aucB = [0.8777, 0.8400, 0.7875, 0.8625, 0.8763, 0.8361]

plt.errorbar(label, aucH, label='Human AUC', alpha=0.7, linewidth = 2, marker = 'o', elinewidth=0.5, capsize=1, capthick=2, ms=11, mec='black',
             color='orangered', zorder=2)
plt.errorbar(label, aucB, label='Bot AUC', alpha=0.7, linewidth = 2, marker = 's', elinewidth=0.5, capsize=1, capthick=2, ms=11, mec='black',
             color='deepskyblue', zorder=2)
plt.ylim(0.73, 0.94)
plt.tick_params(labelsize=16)
plt.xticks([])

plt.legend(fontsize=18)
plt.savefig('gcn_auc.jpg', bbox_inches = 'tight', dpi=500)
plt.show()
```
