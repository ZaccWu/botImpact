```python
import pandas as pd
import re
from collections import Counter
```

```python
user = pd.read_json('Dataset/baidu/user.json')
user.head(5)
```

```python
### Part 1 选择可能能够用到的hashtag
tweet = user['tweet']
analy_user_tweet = tweet[0]
hashtag_extract = [re.findall(r'#(\w+)', line) for line in analy_user_tweet]
hashtags_exist = [word for sublist in hashtag_extract for word in sublist]
print(hashtags_exist)
```

```python
hashtag_list = []
for i in range(len(tweet)):
    analy_user_tweet = tweet[i]
    if analy_user_tweet == None:
        continue
    hashtag_extract = [re.findall(r'#(\w+)', line) for line in analy_user_tweet]
    hashtags = [word for sublist in hashtag_extract for word in sublist]
    hashtag_list.append(hashtags)

all_hashtags = [word for sublist in hashtag_list for word in sublist]
hashtag_freq = Counter(all_hashtags)
```

```python
hashtag_freq_df = pd.DataFrame(hashtag_freq.items(), columns=['word', 'freq']).sort_values(by='freq', ascending=False)
hashtag_freq_df.head()
```

```python
### Part 2 选择讨论过这些话题的推子和对应用户
user_tweet_select = user.copy()
key_hashtags = ['#COVID19', '#coronavirus', '#Covid19', '#Covid_19', '#Coronavirus', '#COVID', '#covid19']




for i in range(len(user)):
    analy_user_tweet = user['tweet'][i]
    if analy_user_tweet == None:
        continue
    #hashtag_extract = [re.findall(r'#(\w+)', line) for line in analy_user_tweet]
    #hashtags_exist = [word for sublist in hashtag_extract for word in sublist]
    select_tweet_list = [sent for sent in analy_user_tweet if any(wd in sent for wd in key_hashtags)]
    #if len(set(hashtags_exist)&set(key_hashtags))>0:
    #    select_uid.append(i)
    user_tweet_select.at[i, 'tweet'] = select_tweet_list
```

```python
select_uid = []
for i in range(len(user_tweet_select)):
    if user_tweet_select['tweet'][i] == None:
        continue
    if len(user_tweet_select['tweet'][i])>0:
        select_uid.append(i)
print(select_uid)
```

```python
select_user = user_tweet_select.loc[select_uid]
bot_label = pd.read_csv('Dataset/baidu/train.csv', sep='\t')
user_with_label = pd.merge(select_user, bot_label, on='ID', how='inner')
user_with_label['Label'].value_counts()
```

```python
# sentiment analysis based on VADER
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
StScore = SentimentIntensityAnalyzer()
StScore_all = []
user_with_label['StScore'] = 0
for i in range(len(user_with_label)):
    user_with_label.loc[i, 'StScore'] = StScore.polarity_scores(user_with_label['tweet'][i])['compound']
```

```python
user_with_label.to_csv('Dataset/baidu/processed/2402-Senti.csv',index=False)
```
