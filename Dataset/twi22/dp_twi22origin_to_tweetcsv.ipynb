```python
import pandas as pd
import ijson
import json
```

```python
# only remain uid and tweet
i = 2
Tweets0 = {'uid': [], 'tweet': []}
p, q = 0, 0
with open('D:/wzy/bot/twi22/tweet_'+str(i)+'.json', 'r') as f:
    parser = ijson.parse(f)
    #key_prefix = ['item.author_id','hashtags','text']
    # for i in range(1000):
    #     prefix, event, value = next(parser)
    for prefix, event, value in parser: # all the file
        if prefix=='item.author_id':
            Tweets0['uid'].append(value)
            p+=1
        if prefix=='item.text':
            Tweets0['tweet'].append(value)
            q+=1
            if q%100==0:
                print(p,q)
Tweets0 = pd.DataFrame(Tweets0)
print(len(Tweets0['uid']), len(Tweets0['tweet']))
Tweets0.to_csv('tweets'+str(i)+'.csv', encoding='utf-8', index=False)
```
